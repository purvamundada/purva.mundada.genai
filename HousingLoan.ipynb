```python
import matplotlib.pyplot as plt
```

```python
def calculate_emi(principal, rate, tenure):
    monthly_rate = rate / (12 * 100)
    emi = (principal * monthly_rate * (1 + monthly_rate) ** tenure) / ((1 + monthly_rate) ** tenure - 1)
    return emi
```

```python
def early_closure_interest_loss(principal, rate, tenure, months_paid):
    emi = calculate_emi(principal, rate, tenure)
    remaining_tenure = tenure - months_paid
    balance = principal - emi * months_paid
    interest_loss = (emi * remaining_tenure) - balance
    return interest_loss
```

```python
principal = 500000
rate = 8.5
tenure = 120
months_paid = 60

emi = calculate_emi(principal, rate, tenure)
print(f"Monthly EMI: {emi:.2f}")

interest_loss = early_closure_interest_loss(principal, rate, tenure, months_paid)
print(f"Interest Lost if Closed Early: {interest_loss:.2f}")

months = list(range(1, tenure + 1))
emi_values = [emi] * tenure
```

```python
plt.plot(months, emi_values)
plt.title("EMI vs Months")
plt.xlabel("Months")
plt.ylabel("EMI Amount")
plt.show()
```
